```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
hover = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Year, Gold)", "(@year, @gold)"),
])
p = figure(title="Olympic Medal Data", x_axis_label="Year", y_axis_label="Gold", width=1000, height=700, tools=[hover])
p.circle("year", "gold", size=10, source=data_frame)
show(p)
```
